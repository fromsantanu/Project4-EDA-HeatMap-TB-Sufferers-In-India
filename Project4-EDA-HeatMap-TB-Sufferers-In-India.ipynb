# HeatMap of TB-Sufferers in Major Cities of India

HeatMap of tuberculosis (TB) sufferers in major cities of India

### Data Generation (Synthetic)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# List of major cities in India
cities = [
    "Mumbai", "Delhi", "Bangalore", "Hyderabad", "Ahmedabad", "Chennai",
    "Kolkata", "Surat", "Pune", "Jaipur", "Lucknow", "Kanpur",
    "Nagpur", "Visakhapatnam", "Bhopal", "Patna", "Ludhiana", "Agra",
    "Nashik", "Faridabad", "Meerut", "Rajkot", "Varanasi", "Srinagar",
    "Aurangabad", "Dhanbad", "Amritsar", "Allahabad", "Ranchi", "Gwalior",
    "Jodhpur", "Madurai", "Raipur", "Kota", "Chandigarh", "Guwahati"
]

# Generate synthetic data for the number of TB sufferers
np.random.seed(42)  # For reproducibility
tb_sufferers = np.random.randint(500, 10000, size=len(cities))

# Create a DataFrame
tb_data = pd.DataFrame({
    "City": cities,
    "TB_Sufferers": tb_sufferers
})

# Define city locations with precise coordinates
city_locations = {
    "Mumbai": (72.8777, 19.0760), "Delhi": (77.1025, 28.7041), "Bangalore": (77.5946, 12.9716),
    "Hyderabad": (78.4867, 17.3850), "Ahmedabad": (72.5714, 23.0225), "Chennai": (80.2707, 13.0827),
    "Kolkata": (88.3639, 22.5726), "Surat": (72.8311, 21.1702), "Pune": (73.8567, 18.5204),
    "Jaipur": (75.7873, 26.9124), "Lucknow": (80.9462, 26.8467), "Kanpur": (80.3319, 26.4499),
    "Nagpur": (79.0882, 21.1458), "Visakhapatnam": (83.2185, 17.6868), "Bhopal": (77.4126, 23.2599),
    "Patna": (85.1376, 25.5941), "Ludhiana": (75.8573, 30.9000), "Agra": (78.0081, 27.1767),
    "Nashik": (73.7898, 19.9975), "Faridabad": (77.3178, 28.4089), "Meerut": (77.7064, 28.9845),
    "Rajkot": (70.8022, 22.3039), "Varanasi": (83.0076, 25.3176), "Srinagar": (74.7973, 34.0837),
    "Aurangabad": (75.3433, 19.8762), "Dhanbad": (86.4304, 23.7957), "Amritsar": (74.8723, 31.6340),
    "Allahabad": (81.8463, 25.4358), "Ranchi": (85.3096, 23.3441), "Gwalior": (78.1791, 26.2183),
    "Jodhpur": (73.0243, 26.2389), "Madurai": (78.1198, 9.9252), "Raipur": (81.6296, 21.2514),
    "Kota": (75.8577, 25.2138), "Chandigarh": (76.7794, 30.7333), "Guwahati": (91.7362, 26.1445)
}

# Add location coordinates to the dataset
tb_data['Longitude'] = tb_data['City'].map(lambda x: city_locations[x][0])
tb_data['Latitude'] = tb_data['City'].map(lambda x: city_locations[x][1])

print(tb_data.head())
```

### Create a Scatter Plot based on Location

```python

# Create the scatter plot
plt.figure(figsize=(14, 10))

# Scatter plot with a color gradient
scatter = plt.scatter(tb_data['Longitude'], tb_data['Latitude'], 
                      c=tb_data['TB_Sufferers'], cmap='OrRd', 
                      s=100, alpha=0.7, edgecolor='k')

# Adding labels to each point
for i, row in tb_data.iterrows():
    plt.text(row['Longitude'], row['Latitude'], row['City'], fontsize=8, ha='right')

# Adding color bar
plt.colorbar(scatter, label='Number of TB Sufferers')

# Title and labels
plt.title('Heatmap of TB Sufferers in Major Indian Cities', fontsize=15)
plt.xlabel('Longitude')
plt.ylabel('Latitude')

plt.grid(True)
plt.show()
```
